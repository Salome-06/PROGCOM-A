```python
class Casa:
  """
  Construir el constructor
  argumentos:
     color(str): color de la casa
     habitaciones(int): el nmero de habitaciones de la casa
     baos(int): el nmero de baos en la casa
     estacionamiento(bool): si tiene o no
     tiene_patio(bool): si tiene patio o no
  """
  def __init__(self, color, habitaciones, baos, estacionamiento, tiene_patio):
    self.color = color
    self.habitaciones = habitaciones
    self.baos = baos
    self.estacionamiento = estacionamiento
    self.tiene_patio = tiene_patio
  def descripcion(self):
    #Describir la casa
    descripcion=f"La casa es de color {self.color}, tiene {self.habitaciones} habitacion(es) y {self.baos} baos"
    if self.estacionamiento:
      descripcion += "  y tiene  estacionamiento."
    if self.tiene_patio:
      descripcion += "  y tiene  patio."
    return descripcion


  def pintar(self,nuevo_color):
    self.color=nuevo_color
    print(f"La casa ha sido pintada de color {self.color}")
```

```python
#Descripcin detallada
class Floristeria(Casa):
  """
  Constructor de la floristera
  Argumentos:
    color(str): color de la floristera
    numero_habitaciones(int): el nmero de habitaciones de la floristera
    numero_baos(int): el nmero de baos de la floristera
    estacionamiento(bool): si tiene o no
    tiene_patio(bool): si tiene patio o no
    tipos_flores(list): Lista de flores
    nombre_floristeria: Nombre de la floristera
    diseo(str): Descripcin del diseo de la floristera.

    """
  def __init__(self,color,numero_habitaciones,numero_baos, estacionamiento, tiene_patio, nombre_floristeria,tipo_flores, diseo):
    super().__init__(color,numero_habitaciones,numero_baos,estacionamiento,tiene_patio)
    self.nombre_floristeria=nombre_floristeria
    self.tipo_flores=tipo_flores
    self.diseo=diseo

  def descripcion(self):
    #Descripcin
        descripcion=f"La floristera se llama {self.nombre_floristeria}. {self.diseo}. Tiene {self.habitaciones} saln(es) y {self.baos} bao(s)."
        if self.estacionamiento:
            descripcion += " Tiene estacionamiento."
        if self.tiene_patio:
            descripcion += " Tiene patio."
        descripcion += f" Estn ofrecidas las siguientes flores: {', '.join(self.tipo_flores)}"
        return descripcion
```
