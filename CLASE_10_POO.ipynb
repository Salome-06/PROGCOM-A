```python
class Casa:
  """
  Construir el constructor
  argumentos:
     color(str): color de la casa
     habitaciones(int): el número de habitaciones de la casa
     baños(int): el número de baños en la casa
     estacionamiento(bool): si tiene o no
     tiene_patio(bool): si tiene patio o no
  """
  def __init__(self, color, habitaciones, baños, estacionamiento, tiene_patio):
    self.color = color
    self.habitaciones = habitaciones
    self.baños = baños
    self.estacionamiento = estacionamiento
    self.tiene_patio = tiene_patio
  def descripcion(self):
    #Describir la casa
    descripcion=f"La casa es de color {self.color}, tiene {self.habitaciones} habitacion(es) y {self.baños} baños"
    if self.estacionamiento:
      descripcion += "  y tiene  estacionamiento."
    if self.tiene_patio:
      descripcion += "  y tiene  patio."
    return descripcion


  def pintar(self,nuevo_color):
    self.color=nuevo_color
    print(f"La casa ha sido pintada de color {self.color}")
```

```python
mi_casa=Casa("amarilla", 3,2,True,True)
type(mi_casa)
```

```python
print(mi_casa.descripcion())
```

```python
mi_casa.tiene_patio=False
print(mi_casa.descripcion())
```

```python
#Descripción detallada
class Floristeria(Casa):
  """
  Constructor de la floristería
  Argumentos:
    color(str): color de la floristería
    numero_habitaciones(int): el número de habitaciones de la floristería
    numero_baños(int): el número de baños de la floristería
    estacionamiento(bool): si tiene o no
    tiene_patio(bool): si tiene patio o no
    tipos_flores(list): Lista de flores
    nombre_floristeria: Nombre de la floristería
    diseño(str): Descripción del diseño de la floristería.

    """
  def __init__(self,color,numero_habitaciones,numero_baños, estacionamiento, tiene_patio, nombre_floristeria,tipo_flores, diseño):
    super().__init__(color,numero_habitaciones,numero_baños,estacionamiento,tiene_patio)
    self.nombre_floristeria=nombre_floristeria
    self.tipo_flores=tipo_flores
    self.diseño=diseño

  def descripcion(self):
    #Descripción
        descripcion=f"La floristería se llama {self.nombre_floristeria}. {self.diseño}. Tiene {self.habitaciones} salón(es) y {self.baños} baño(s)."
        if self.estacionamiento:
            descripcion += " Tiene estacionamiento."
        if self.tiene_patio:
            descripcion += " Tiene patio."
        descripcion += f" Están ofrecidas las siguientes flores: {', '.join(self.tipo_flores)}"
        return descripcion
```

```python
mi_floristeria=Floristeria("azul", 1, 1, True, False, "FLOWERS_SHINE", ["girasoles", "orquídeas", "tulipanes", "rosas", "claveles", "rampión"], "tiene un diseño floral al rededor de toda la estructura, o sea las paredes exteriores")
mi_floristeria.descripcion()
```
