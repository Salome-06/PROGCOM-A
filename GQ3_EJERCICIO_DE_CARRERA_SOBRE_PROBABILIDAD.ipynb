```python
#EJERCICIO DE CARRERA SOBRE PROBABILIDAD
#Mensaje de bienvenida y explicación
print("¡Bienvenido al programa de solución de problemas de probabilidad!")
print("En este programa, resolverás un problema de probabilidad basado en la estadística de cursos.")
print("A continuación, podrás seleccionar qué cálculo deseas realizar.")
print("---")

print("--- Problema de Probabilidad ---")
print("Cálculo de probabilidades en base a una encuesta de 110 estudiantes.")
print("---")

#Datos del problema
total_estudiantes = 110
inglés = 75
historia = 52
matemáticas = 50

#Intersecciones
ingles_historia = 33
ingles_matematicas = 30
historia_matematicas = 22
ingles_historia_matematicas = 13

#Bucle principal para repetir el menú
while True:
    #Menú de opciones
    print("\nElige el punto que deseas calcular:")
    print("a) Probabilidad de tomar Inglés")
    print("b) Probabilidad de tomar Historia o Matemáticas")
    print("c) Probabilidad de tomar solamente Matemáticas")
    print("d) Probabilidad de tomar Inglés e Historia, pero no Matemáticas")
    print("e) Probabilidad de no tomar ningún curso")
    print("f) Probabilidad de tomar exactamente dos cursos")
    print("q) Salir del programa")
    print("---")

    opcion = input("Ingresa la letra del punto (a-f) o 'q' para salir: ").lower()

    #Lógica para salir del bucle
    if opcion == 'q':
        print("Saliendo del programa. ¡Hasta luego!")
        break  #Hace que termine el bucle y el programa

    #Lógica para calcular y mostrar la respuesta según la opción elegida
    elif opcion == 'a':
        probabilidad_inglés = inglés / total_estudiantes
        print(f"\na) Probabilidad de tomar Inglés: {probabilidad_inglés:.4f}")
    elif opcion == 'b':
        probabilidad_historia_o_matematicas = (historia + matemáticas - historia_matematicas) / total_estudiantes
        print(f"\nb) Probabilidad de tomar Historia o Matemáticas: {probabilidad_historia_o_matematicas:.4f}")
    elif opcion == 'c':
        solo_matematicas = matemáticas - (ingles_matematicas - ingles_historia_matematicas) - (historia_matematicas - ingles_historia_matematicas) - ingles_historia_matematicas
        probabilidad_solo_matematicas = solo_matematicas / total_estudiantes
        print(f"\nc) Probabilidad de tomar solamente Matemáticas: {probabilidad_solo_matematicas:.4f}")
    elif opcion == 'd':
        ingles_historia_no_matematicas = ingles_historia - ingles_historia_matematicas
        probabilidad_ingles_historia_no_matematicas = ingles_historia_no_matematicas / total_estudiantes
        print(f"\nd) Probabilidad de tomar Inglés e Historia, pero no Matemáticas: {probabilidad_ingles_historia_no_matematicas:.4f}")
    elif opcion == 'e':
        union_tres_cursos = (inglés + historia + matemáticas) - (ingles_historia + ingles_matematicas + historia_matematicas) + ingles_historia_matematicas
        no_toma_ninguno = total_estudiantes - union_tres_cursos
        probabilidad_ninguno = no_toma_ninguno / total_estudiantes
        print(f"\ne) Probabilidad de no tomar ningún curso: {probabilidad_ninguno:.4f}")
    elif opcion == 'f':
        solo_dos_cursos = (ingles_historia - ingles_historia_matematicas) + (ingles_matematicas - ingles_historia_matematicas) + (historia_matematicas - ingles_historia_matematicas)
        probabilidad_dos_cursos = solo_dos_cursos / total_estudiantes
        print(f"\nf) Probabilidad de tomar exactamente dos cursos: {probabilidad_dos_cursos:.4f}")
    else:
        print("\nOpción no válida. Por favor, ingresa una letra de la 'a' a la 'f' o 'q' para salir.")
```
